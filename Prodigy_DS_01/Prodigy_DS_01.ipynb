```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/heart.csv')

# Set the style for seaborn
sns.set(style="whitegrid")
```

```python
# 1. Age Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Gender Distribution (Bar Chart)
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Chest Pain Type (cp) Distribution (Bar Chart)
plt.figure(figsize=(10, 6))
sns.countplot(x='cp', data=df)
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Cholesterol Levels Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['chol'], bins=20, kde=True)
plt.title('Cholesterol Levels Distribution')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Resting Blood Pressure (trestbps) Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['trestbps'], bins=20, kde=True)
plt.title('Resting Blood Pressure Distribution')
plt.xlabel('Resting Blood Pressure (mm Hg)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 6. Max Heart Rate (thalach) Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['thalach'], bins=20, kde=True)
plt.title('Max Heart Rate Distribution')
plt.xlabel('Max Heart Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# 7. Number of Major Vessels (ca) Distribution (Bar Chart)
plt.figure(figsize=(10, 6))
sns.countplot(x='ca', data=df)
plt.title('Number of Major Vessels Distribution')
plt.xlabel('Number of Major Vessels (0-3)')
plt.ylabel('Count')
plt.show()
```

```python
# 8. Target Variable Distribution (Bar Chart)
plt.figure(figsize=(8, 6))
sns.countplot(x='target', data=df)
plt.title('Heart Disease Distribution')
plt.xlabel('Target (0 = No Disease, 1 = Disease)')
plt.ylabel('Count')
plt.show()
```

```python
# 9. Categorical Variables Comparison by Gender (Grouped Bar Chart)
categorical_vars = ['cp', 'restecg', 'thal']
for var in categorical_vars:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=var, hue='sex', data=df)
    plt.title(f'{var} Distribution by Gender')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.legend(title='Gender', labels=['Female', 'Male'])
    plt.show()
```
