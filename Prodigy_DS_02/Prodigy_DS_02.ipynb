```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('/content/train.csv')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe().include='all'
```

```python
# Example: Fill missing ages with the median age
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
# Survival Rate
df['Survived'].value_counts().plot(kind='bar')
plt.title('Survival Rate')
plt.show()
```

```python
# Age Distribution
df['Age'].hist()
plt.title('Age Distribution')
plt.show()
```

```python
# Gender and Survival
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Class and Survival
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Class')
plt.show()
```

```python
# Embarked Analysis
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rate by Embarked')
plt.show()
```

```python
# Survival by Age and Class
sns.scatterplot(x='Age', y='Survived', hue='Pclass', data=df)
plt.title('Survival by Age and Class')
plt.show()
```

```python
# Survival by Gender and Class
sns.catplot(x='Pclass', hue='Sex', col='Survived', data=df, kind='count')
plt.title('Survival by Gender and Class')
plt.show()
```
